Data Processing

```python
import os
import re
import pandas as pd
import numpy as np
from pathlib import Path
```

Path in which the notebook is executing

```python
base_path = str(Path(os.path.abspath(os.getcwd())))
files = os.listdir(base_path+"/data/")
files = files[0:6]
```

Load the data and concatenate it

```python
frames = []
for file in files:
     mobility_csv = base_path + "/data/" + file
     mob_df = pd.read_csv(mobility_csv, sep=";")
     frames.append(mob_df)
df = pd.concat(frames)     
```

Drop non relevant columns

```python
df.drop(["Mobility Duration", "Education Level", "Special Needs", "Fewer Opportunities", "GroupLeader",
                     "Sending Organisation Erasmus Code", "Receiving Organisation Erasmus Code"], axis=1, inplace=True)
```

Filter on rows that are students(Learner)

```python
df = df[df["Participant Profile"] == "Learner"]
df
```
