Data Processing

```python
import os
import re
import pandas as pd
import numpy as np
from pathlib import Path
```

Path in which the notebook is executing

```python
base_path = str(Path(os.path.abspath(os.getcwd())))
files = os.listdir(base_path+"/data/")
files = files[3:6]
files
```

Load the data and concatenate it

```python
frames = []
for file in files:
     mobility_csv = base_path + "/data/" + file
     mob_df = pd.read_csv(mobility_csv, sep=";")
     frames.append(mob_df)
df = pd.concat(frames)     
```

Drop non relevant columns

```python
df.drop(["Mobility Duration", "Education Level", "Special Needs", "Fewer Opportunities", "GroupLeader",
                     "Sending Organisation Erasmus Code", "Receiving Organisation Erasmus Code"], axis=1, inplace=True)
```

Filter on rows that are students

```python
df = df[df["Participant Profile"] == "Learner"]
df = df[df["Activity (mob)"].str.contains("Student")]
df
```

```python
df = df.loc[:, ["Sending Organization", "Receiving Organization", "Participants", "Sending Country Code", "Receiving Country Code"]]
df["Sending Organization"] = df["Sending Organization"].str.upper()
df["Receiving Organization"] = df["Receiving Organization"].str.upper()
df["Sending Country Code"] = df["Sending Country Code"].str.upper()
df["Receiving Country Code"] = df["Receiving Country Code"].str.upper()
df
```
